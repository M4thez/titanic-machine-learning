```python
from sklearn.model_selection import train_test_split
from sklearn import tree
import pandas as pd
import numpy as np
import os
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
train_set_survivors = pd.read_csv('train.csv')

train_set = train_set.drop(columns=['Survived'])
train_set_survivors = train_set_survivors.drop(columns=['Age','SibSp','Parch','Name','Ticket', 'Fare', 'Cabin', 'Pclass', 'Sex', 'Embarked', 'PassengerId'])


def data_prep(set):
    set = set.drop(columns=['Name','Ticket', 'Fare', 'Cabin'])
    set = set.fillna(0)
    set['Sex'] = set['Sex'].replace('male', 1)
    set['Sex'] = set['Sex'].replace('female', 0)
    set['Embarked'] = set['Embarked'].replace('C', 0)
    set['Embarked'] = set['Embarked'].replace('S', 1)
    set['Embarked'] = set['Embarked'].replace('Q', 2)
    set['Age'] = set['Age'].astype(np.int64)
    return set
```

```python
data_prep(train_set)
```
